```python
products={
        "Oranges (packaged)": 114.99, 
        "Candy (Rotfront)": 280.00,
        "Boiled sausage": 199.99,
        "Juice J7 (orange)": 119.99,
        "Trout (Seven Seas)": 399.99
        }
stocks = {
        "Boiled sausage": "33%",
        "Juice J7 (orange)": "12%",
        "Trout (Seven Seas)": "18%"
    }


def apply_discounts(products, stocks):
    """
    Напишите функцию apply_discounts(), которая снижает цену продуктов в словаре products на указанный в словаре stocks процент. 
    Функция должна вернуть результирующий словарь, ключи которого — товары, а значения — новые цены.
    Если продукта из словаря stocks нет в словаре products, то его необходимо пропустить. Цены округлите до второго знака после запятой.
    """
    result  = products
    for key, value in stocks.items():
         proc = value[:-1]
         cost = products[key]
         print(key, products[key], float(proc), cost, result[key], round(cost - cost/100 * float(proc), 2))
         result[key] =  round(cost - cost/100 * float(proc), 2)
        
    return result
        


new_products = apply_discounts(products, stocks)
print(new_products)
```

```python
import pandas as pd 

solar_data = pd.read_csv('data/solarpower.csv')
prev_ = 0
next_ = 0
for index, row in solar_data.iterrows():
    curent_value =  row['cum_power']
    if row['cum_power'] == -1:
         prev_= index - 1
         next_= index + 1
         solar_data.at[index,'cum_power']= (solar_data.iloc[prev_]['cum_power'] + solar_data.iloc[next_]['cum_power'])/2
         
#solar_data = solar_data[solar_data['cum_power']==-1] 
print(round(solar_data['cum_power'].mean(),0))
```

```python
prev_ = 0
next_ = 0
solar_data['day_power'] = 0

for index, row in solar_data.iterrows():
    if index == 0:
        solar_data.at[index,'day_power'] = 0
    else:
        prev_= index - 1
        solar_data.at[index,'day_power']= (solar_data.iloc[index]['cum_power'] - solar_data.iloc[prev_]['cum_power'])

solar_data.head()
```

```python
solar_data['date'] = pd.to_datetime(solar_data['date']) 
solar_data['aYear'] = pd.DatetimeIndex( pd.to_datetime(solar_data['date']) ).year
solar_data['aMonth'] = pd.DatetimeIndex( pd.to_datetime(solar_data['date']) ).month
solar_data.info()
```

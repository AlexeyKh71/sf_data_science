### Гипотезы

H0 - Средний размер мидий одинаков в различных городах

Ha - Средний размер мидий отличается в ьзависимости от города производителя 

 ### Объединяем данные в Dataframe
 Использем данные по мидиям из раздела ANOVA-тест

```python
petersburg = [0.0974, 0.1352, 0.0817, 0.1016, 0.0968, 0.1064, 0.105]
# для создания dataframe количество данных в колонках должно быть одинаковым, по этом удаляем посмледнее 
# значение из листа magadan = [0.1033, 0.0915, 0.0781, 0.0685, 0.0677, 0.0697, 0.0764, 0.0689]
magadan = [0.1033, 0.0915, 0.0781, 0.0685, 0.0677, 0.0697, 0.0764]

data = {'setersburg': petersburg,
        'magadan': magadan}

mussels_df = pd.DataFrame(data)

# Установим уровень значимости
alpha = 0.05
```

### Проверка на нормальность

```python
H0 = 'Данные распределены нормально'
Ha = 'Данные не распределены нормально (мы отвергаем H0)'
```

```python
_, p = shapiro(mussels_df)
print('p=%.3f' % p)

# Интерпретация 

if p > alpha:
	print(H0)
else:
	print(Ha)
```

### Tест на корреляцию 
Однофакторный дисперсионный анализ ANOVA проверяет нулевую гипотезу о том, что две или более групп имеют одинаковое среднее значение генеральной совокупности.

```python
# заносим вводы по теории в переменные
H0 = 'Нет значимой разницы между средним размером раковины мидий в двух разных местах.'
Ha = 'Есть значимая разница между средним размером раковины мидий в двух разных местах.'

# вызов процедуры теста на корреляцию
_, p = f_oneway(petersburg, magadan)

# вывод тетха о результатах теста на коррелицию
if p>alpha:
  print(f"{p} > {alpha}. Мы не можем отвергнуть нулевую гипотезу. {H0}")
else:
  print(f"{p} <= {alpha}. Мы отвергаем нулевую гипотезу. {Ha}")
```

### Сравнительный тест
Зависимая переменная (диаметр мидии) является количественной. Группы происходят из разных совокупностей. Следовательно, мы используем независимый T-тест.

```python
# заносим вводы по теории в переменные
H0 = 'Нет значимой разницы между диаметрами мидий в разных городах.'
Ha = 'Есть значимая разница между диаметрами мидий в разных городах.'

# процедура выполнен сравнительного теста
def t_test_1(df):
    print('\n' + "*** Результаты независимого T-теста ***")
    test_results = ttest_ind(mussels_df['setersburg'], mussels_df['magadan'], equal_var=True)

    p = round(test_results[1],2)

    if p>alpha:
        print(f"{p} > {alpha}. Мы не можем отвергнуть нулевую гипотезу. {H0}")
    else:
        print(f"{p} <= {alpha}. Мы отвергаем нулевую гипотезу. {Ha}")

t_test_1(mussels_df)
```
